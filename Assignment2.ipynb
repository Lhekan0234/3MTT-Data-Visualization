```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load corrupted image
img = cv2.imread("/content/students original.jpg") # bgr (blue, green, red)
```

```python
plt.imshow(img)
```

```python
alpha = 1
# Contrast factor (try values 1.0 3.0)
beta = 30
#Brightness shift (try values 100 to +100)
restored = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
```

```python
plt.imshow(restored)
```
